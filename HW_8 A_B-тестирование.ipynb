### 2. Если мы увеличиваем объем выборки в два раза (при условии, что показатель стандартного отклонения остается неизменным), то 95% доверительный интервал

Выберите один вариант из списка

1. ***стал более узким***
```
в формуле расчета сигмы в знаменателе корень из n => при увеличении n, стандартное отклонение будет уменьшаться. Доверительный интервал рассчтывается как (мат.ожидание -+сигма) поэтому с уменьшением сигмы интервал становится уже
```
2. возможны оба варианта
3. стал более широким

````python
# Проверка
import statsmodels.api as sm

k = 479

for n in [1000, 2000]:
    print(
        f'доверительный интервал, при n={n} ',
        sm.stats.proportion_confint(count=k, nobs=n, alpha=(1 - 0.95), method='normal'))
````

### 3. В центре 95% доверительного интервала, рассчитанного по выборочным значениям, находится:

Выберите один вариант из списка

1. Значение стандартной ошибки среднего
2. **Выборочное среднее значение**
3. Среднее значение генеральной совокупности

````python
# Проверка
k, n = 479, 1000

interval = sm.stats.proportion_confint(k, n, alpha=0.05, method='normal')
np.median(interval) == (k / n)
````
